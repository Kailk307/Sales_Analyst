## Task 2: Clean Data

```python
# Kiểm tra giá trị thiếu
print(df.isnull().sum())

# Loại bỏ các giá trị thiếu (nếu có)
df = df.dropna()

# Đảm bảo cột ngày/thời gian ở đúng định dạng
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])

# Tạo cột tháng, quý, năm từ ORDERDATE
df['Month'] = df['ORDERDATE'].dt.month
df['Quarter'] = df['ORDERDATE'].dt.quarter
df['Year'] = df['ORDERDATE'].dt.year
```

```python
total_sales = df['SALES'].sum()
print(f"Tổng doanh thu: {total_sales}")
```

```python
# Doanh thu theo tháng
monthly_sales = df.groupby(['Year', 'Month'])['SALES'].sum().reset_index()

# Doanh thu theo quý
quarterly_sales = df.groupby(['Year', 'Quarter'])['SALES'].sum().reset_index()

print(monthly_sales)
print(quarterly_sales)

```

```python
productline_sales = df.groupby('PRODUCTLINE')['SALES'].sum().sort_values(ascending=False)
print(productline_sales)
```

```python
df_sorted = df.sort_values('SALES', ascending=False)
top_customers = (
    df_sorted.drop_duplicates(subset='CUSTOMERNAME')
    .head(10)
)
print(top_customers)

```

```python
unique_df = df.drop_duplicates(subset=['COUNTRY', 'SALES'])
country_sales = (
    unique_df.groupby('COUNTRY')['SALES']
    .sum()
    .sort_values(ascending=False)
)
print(country_sales)

```

```python
top_products = df.groupby('PRODUCTLINE')['QUANTITYORDERED'].sum().sort_values(ascending=False)
print(top_products)

```
